## POPULARITY BASED RECOMMENDER SYSTEM

```python
# Merging ratings and books
ratings_with_books = ratings.merge(books, on='ISBN')
ratings_with_books.head()
```

```python
num_ratings_df = ratings_with_books.groupby("Book-Title").count()["Book-Rating"].reset_index()
num_ratings_df.rename(columns={"Book-Rating": "num_rating"}, inplace=True)
num_ratings_df
```

```python
avg_ratings_df = ratings_with_books.groupby("Book-Title")["Book-Rating"].mean()
avg_ratings_df = avg_ratings_df.rename("avg_rating").reset_index()
avg_ratings_df.head()
```

```python
popular_df = num_ratings_df.merge(avg_ratings_df, on="Book-Title")
popular_df.head()
```

```python
popular_df = popular_df[popular_df['num_rating']>250].sort_values("avg_rating", ascending= False).head(50)
popular_df.head()
```

```python
popular_df = popular_df.merge(books, on="Book-Title").drop_duplicates("Book-Title")[["Book-Title", "Book-Author", "Image-URL-M", "num_rating", "avg_rating"]]
popular_df.head()
```

## COLLABORATIVE FILTERING BASED RECOMMENDATION SYSTEM

```python
# Rating on the basis of users
x = ratings_with_books.groupby("User-ID").count()["Book-Rating"]>200
educated_users = x[x].index
```

```python
filtered_rating = ratings_with_books[ratings_with_books["User-ID"].isin(educated_users)]
```

```python
# Rating on the basis of books
y = filtered_rating.groupby("Book-Title").count()["Book-Rating"]>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating["Book-Title"].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index="Book-Title", columns="User-ID", values="Book-Rating")
pt.fillna(0, inplace=True)
pt.head()
```

```python
# Module for calculating distance
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
def recommend(book_name):
    # Index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x: x[1], reverse=True)[1:6]
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend("1984")
```

```python
recommend("Message in a Bottle")
```
